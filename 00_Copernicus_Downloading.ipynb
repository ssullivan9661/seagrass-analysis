# Sentinel-2 Download via Copernicus

Sentinel-2 satelliete data was collected via [Copernicus Data Space Ecosystem](https://dataspace.copernicus.eu/data-collections/copernicus-sentinel-missions/sentinel-2). Downloaded for the years 2015-2024, around the AOI: Upper Keys Florida particularly around John Pennekamp Coral Reef State Park.

1. Copernicus
2. Create AOI - using coordinates of bounding box in JSON
3. `Search` Tab Sentinel-2 Data
    - Type: 
        L1C for 2015-2016
        L1A for 2017-2024
        - Filters: **L2A is checked**

Note: for the fastest loading and most effecient way to add the scenes to workspace I found was to search years in groups of two.



### Imports

```python
import geopandas as gpd
from shapely.geometry import box
```

## Coordinates of AOI

```python
# Load shapefile
gdf = gpd.read_file('C:/Users/sulli/DATA_Science/seagrass_project/study_area/UpperKeys_FL.shp')

# Ensure correct CRS
if gdf.crs != "EPSG:4326":
    gdf = gdf.to_crs("EPSG:4326")

# Get bounding box
minx, miny, maxx, maxy = gdf.total_bounds

# Create bbox polygon (optional)
bbox_poly = box(minx, miny, maxx, maxy)
bbox_gdf = gpd.GeoDataFrame(geometry=[bbox_poly], crs="EPSG:4326")

print(minx, miny, maxx, maxy)
```

### JSON to create AOI in Copernicus

```python
# Reduce number of vertices (tolerance in degrees, e.g., 0.001 ~ 100 m)
gdf_simplified = gdf.copy()
gdf_simplified['geometry'] = gdf_simplified.geometry.simplify(tolerance=0.001)

# Convert to GeoJSON
aoi_geometry_simplified = gdf_simplified.geometry.iloc[0].__geo_interface__

import json
aoi_json_simplified = json.dumps(aoi_geometry_simplified, indent=2)
print(aoi_json_simplified)
```
